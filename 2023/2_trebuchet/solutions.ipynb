In this example, the calibration values are 29, 83, 13, 24, 42, 14, and 76. Adding these together produces 281.

What is the sum of all of the calibration values?

## Consider your entire calibration document. What is the sum of all of the calibration values?

### Step 0: Read the inputs

```python
INPUT_FILE = "inputs.txt"
with open(INPUT_FILE, "r") as f: inputs = f.read().splitlines()
inputs
```

### Step 1: Extract all integers from the strings

#### Step 1.1 Create regex for the strings

```python
import regex as re

mp = {"one": "1", "two": "2", "three": "3", "four": "4", "five": "5", "six": "6", "seven": "7", "eight": "8", "nine": "9"}
pattern = "[0-9]|(?:" + "|".join(map(re.escape, mp.keys())) + ")"
pattern
```

#### Step 1.2 Find all numbers in the input

```python
integers = []

for i in inputs:
    current_list = re.findall(pattern, i, overlapped=True) # note: overlapped=True
    print(current_list)
    for index, text in enumerate(current_list):
        if text in mp: current_list[index] = mp[text]
    integers.append(current_list)

integers
```

### Step 2: Take all first and last digits of each list and paste them together as string

```python
results = []
for o in integers: results.append(int(o[0] + o[-1]))
results
```

### Step 3: Sum the results: Finish!

```python
sum(results)
```
